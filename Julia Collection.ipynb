```python
#Pic1
import matplotlib.pyplot as plt
import random as rand
import numpy as np

def func(z):
    f = [0,0]
    c = (-0.4 , -0.6)
    f[0] = z[0]**2  - z[1]**2 +c[0]
    f[1] = 2*z[0]*z[1] + c[1]
    return f

###main Block
N =128
s = 3
color = []
for c in range(16):
      color.append( (rand.random() ,rand.random(),rand.random()) )

x = np.linspace(-1 , 1 ,N) 
y = np.linspace(-1,1,N) 

for i in range(N):
    for j in range (N):
        p = (x[i] , y[j])
        for l in range(5): # to reduse the error
            p = func(p)
            if l==4:
                distance = np.sqrt(p[0]**2 + p[1]**2)
                if  distance <0.1 :
                    plt.scatter(x[i], y[j] ,s  , [color[0]])
                elif  (distance > 0.1) and(distance < 0.2) :
                    plt.scatter(x[i], y[j] ,s  , [color[1]])
                elif  (distance > 0.2) and(distance < 0.3) :
                    plt.scatter(x[i], y[j] ,s  , [color[2]])
                elif  (distance > 0.3) and(distance < 0.4) :
                    plt.scatter(x[i], y[j] ,s  , [color[3]])
                elif  (distance > 0.4) and(distance < 0.5) :
                    plt.scatter(x[i], y[j] ,s  , [color[4]])
                elif  (distance > 0.5) and(distance < 0.6) :
                    plt.scatter(x[i], y[j] ,s  , [color[5]])    
                elif  (distance > 0.6) and(distance < 0.7) :
                    plt.scatter(x[i], y[j] ,s  , [color[6]])
                elif  (distance > 0.7) and(distance < 0.8) :
                    plt.scatter(x[i], y[j] ,s  , [color[7]])
                elif  (distance > 0.8) and(distance < 0.9) :
                    plt.scatter(x[i], y[j] ,s  , [color[8]])
                elif  (distance > 0.9) and(distance < 1) :
                    plt.scatter(x[i], y[j] ,s  , [color[9]])
                elif  (distance > 1) and(distance < 1.1) :
                    plt.scatter(x[i], y[j] ,s  , [color[10]])
                elif  (distance > 1.1) and(distance < 1.3) :
                    plt.scatter(x[i], y[j] ,s  , [color[11]])
                elif  (distance > 1.3) and(distance < 1.5) :
                    plt.scatter(x[i], y[j] ,s  , [color[12]])
                elif  (distance > 1.5) and(distance < 1.7) :
                    plt.scatter(x[i], y[j] ,s  , [color[13]])
                elif  (distance > 1.7) and(distance < 2) :
                    plt.scatter(x[i], y[j] ,s  , [color[14]])    
                else:
                    plt.scatter(x[i], y[j] ,s  , [color[15]])

plt.show()                    
        
```
