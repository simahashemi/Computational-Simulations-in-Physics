```python
from scipy import sqrt, zeros
import matplotlib.pyplot as plt
from random import random, randint
import numpy as np

def midpoint(point1, point2):
    middle = ((point1[0] + point2[0])* 0.5 , (point1[1] + point2[1]) *0.5)
    return middle

corner = [(0, 0), (0.5, np.cos(np.pi/3)), (1, 0)]

N = 4000
x = random()
y = random()
for i in range(1, N):
    for j in range(20):
        k = randint(0, 2)
        x, y = midpoint( corner[k], (x, y) )
    plt.scatter(x, y ,s = 1 , c = 'r')

plt.show()
```
