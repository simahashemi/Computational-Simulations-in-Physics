```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
plt.axis('equal')
ax.set(xlabel='', ylabel='',
       title='Sierpinski triangle')
corner = [(0,0) , (0.5,np.sin(np.pi/3)) , (1,0)]
length = 1

def drawtriangle (p1, p2 , p3 ):
    X = [p1[0] , p2[0] , p3[0]]
    Y = [p1[1] , p2[1] , p3[1]]
    ax.fill(X, Y, 'g')
    return
    
def midpoint(point1, point2):
    middle = ((point1[0] + point2[0])* 0.5 , (point1[1] + point2[1]) *0.5)
    return middle    

 
def sierpinski (level , point , length):
    if level == 0 :
        drawtriangle(point , (point[0]+length/2 , point[1]+ length*np.sin(np.pi/3)) , (point[0]+length , point[1]))
    else:
        sierpinski(level-1 , midpoint(point , (0,0)) , length/2)
        sierpinski(level-1 , midpoint(point , corner[1]) , length/2)
        sierpinski(level-1 , midpoint(point , corner[2]) , length/2)
        
    return  
            
###main block        
                 
sierpinski (1 , corner[0] , length)        
    
plt.show()
        
```
